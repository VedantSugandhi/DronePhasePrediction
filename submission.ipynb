```python
import pandas as pd
# First load the dataset into pandas dataframe
dataset = pd.read_csv('training-set.csv',delimiter=',')
```

```python
dataset.head()
```

```python
#For calculating total no of rows
rows = dataset.count()[0]
print("total number of rows =", rows)
#For calculating total no of cols
cols = len(dataset.columns)
print("total number of columns =",cols)
# For calculating Number of columns having non-numeric values
non_numeric_columns = dataset.select_dtypes(exclude=['number'])
num = len(non_numeric_columns.columns)
print(f"Number of columns having non-numeric values= {num}")
# For calculating Number of missing values
missing_values = dataset.isnull().sum()
missing_values = missing_values.sort_values(ascending=False)
print("Number of missing values in each column:")
missing_columns = [(column, missing_number) for column, missing_number in missing_values.items() if missing_number>0]
missing_columns.sort(key=lambda x: x[1], reverse=True)
summary =   f"total number of rows = {rows}\n"\
                f"total number of columns = {cols}\n"\
                f"number of columns having non-numeric values = {non_numeric_columns}\n"\
                f"columns with missing values = {missing_columns}\n"
print(summary)
```

```python
target_column = 'Operating_Phase'
x = dataset.drop(columns=[target_column])  # X contains all columns except the target column
y = dataset[target_column]
print(x)
```

```python
from sklearn.model_selection import train_test_split
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y,train_size=0.75, test_size=0.25, random_state=12346)

columns_to_drop = ['GPS:dateTimeStamp']

X_train = X_train.drop(columns=columns_to_drop)
X_test = X_test.drop(columns=columns_to_drop)

print("X Train Data")
print(X_train)
print("X Test Data")
print(X_test)
```

```python
categorical_columns = X_train.select_dtypes(exclude=['number']).columns
X_train_ohe = pd.get_dummies(X_train,  columns=categorical_columns, dtype=int)
X_test_ohe = pd.get_dummies(X_test,  columns=categorical_columns, dtype=int)
print(categorical_columns)
```

```python
from sklearn.preprocessing import MinMaxScaler
#X_scaled = (X - X.min()) / (X.max() - X.min())
# Create an instance of MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled_mm = scaler.fit_transform(X_train_ohe)
X_test_scaled_mm = scaler.transform(X_test_ohe)
print(X_test_scaled_mm)
```

```python
missing_values = X_train_ohe.isnull().sum()
missing_counts = pd.DataFrame({'variable_name': missing_values.index, 'num_of_missing_values': missing_values.values})
X_train_ohe_imputed = X_train_ohe.fillna(X_train_ohe.mean())
X_test_ohe_imputed = X_test_ohe.fillna(X_test_ohe.mean())
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer instance
imputer = SimpleImputer(strategy='median')  # You can choose a different strategy like 'median' or 'most_frequent'

# Fit and transform the imputer on your feature matrix
X_train_imputed = imputer.fit_transform(X_train_ohe_imputed)
X_test_imputed = imputer.transform(X_test_ohe_imputed)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.fit_transform(y_test)

# # Drop rows with missing values (NaN)
# X_train = X_train.dropna()
# y_train_encoded = y_train[X_train.index]  # Adjust the target variable accordingly

# # Similarly, drop rows with missing values in the test data
# X_test = X_test.dropna()
# y_test_encoded = y_test[X_test.index]  # Adjust the target variable accordingly

correlation_matrix = X_train_ohe_imputed.corrwith(pd.Series(y_train_encoded))
print(y_train_encoded)

correlation_matrix.dropna(inplace=True)
top75_corr = correlation_matrix.abs()#.nlargest(70)
top75_df = pd.DataFrame({'variable': top75_corr.index, 'corr_score': top75_corr.values})
print(top75_df)
```

```python
top75_variables = top75_df['variable'].tolist()

X_train_t75 = X_train_ohe_imputed[top75_variables]
X_test_t75 = X_test_ohe_imputed[top75_variables]
print(len(top75_variables))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
rf_classifier.fit(X_train_imputed, y_train_encoded)

y_pred = rf_classifier.predict(X_test_imputed)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Evaluate the model
accuracy = accuracy_score(y_test_encoded, y_pred)
report = classification_report(y_test_encoded, y_pred)
conf_matrix = confusion_matrix(y_test_encoded, y_pred)

print(f"Accuracy: {accuracy}")
print(report)
print("Confusion Matrix:")
print(conf_matrix)
```
